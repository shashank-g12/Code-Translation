```python
import os

def get_file_content_length(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            return len(content)
    except FileNotFoundError:
        return -1  # Return -1 if the file is not found

def select_files_with_smallest_content(directory, num_files=10):
    # Get all files in the directory
    all_files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

    # Sort files based on content length
    sorted_files = sorted(all_files, key=lambda f: get_file_content_length(os.path.join(directory, f)))

    # Select the first 'num_files' files with smallest content lengths
    selected_files = sorted_files[:num_files]

    return selected_files

# Example usage
directory_path = "D:\\MTP_project\\codes\\PLTranslation_data\\dataset\\avatar\Java\\Code"
selected_files = select_files_with_smallest_content(directory_path)

# Print the selected files
print("Selected Files:")
for file_name in selected_files:
    print(file_name)
'''
output-
codeforces_678_A.java
codeforces_379_A.java
atcoder_ABC168_B.java
codeforces_23_B.java
codeforces_569_A.java
atcoder_ABC172_D.java
codeforces_581_A.java
atcoder_ABC178_A.java
atcoder_ABC125_A.java
codeforces_110_B.java
'''
```
